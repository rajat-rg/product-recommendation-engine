```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("flipkart.csv")
data.head()
```

```python
data.info()
```

```python
data.groupby("product_rating")['uniq_id'].count()
```

```python
data = data[['uniq_id','product_url','product_name','product_category_tree','image','description','brand']]
data.head()
```

```python
data.isnull().sum()
```

```python
data['image'] = data['image'].fillna("['#']")
data['description'] = data['description'].fillna("['']")
data['brand'] = data['brand'].fillna("['']")
# data['product_specifications'] = data['product_specifications'].fillna("")
data.isnull().sum()
```

```python
import ast
def convert(obj):
    i = ast.literal_eval(obj)
    return i[0]
```

```python
data['image'] = data['image'].apply(convert)
data['product_category_tree'] = data['product_category_tree'].apply(convert)
# data.head()
```

```python
data['brand']= data['brand'].apply(lambda x: x.split())
```

```python
def convert2(string):
    li = list(string.split(" "))
    return li

data['description'] = data['description'].apply(convert2) 
```

```python
data['product_category_tree'] = data['product_category_tree'].apply(lambda x: x.split(">>"))
```

```python
data['product_category_tree'] = data['product_category_tree'].apply(lambda x: [i.replace(" ","") for i in x])
data['brand'] = data['brand'].apply(lambda x: [i.replace(" ","") for i in x])
data['description'] = data['description'].apply(lambda x: [i.replace(" ","") for i in x])
```

# creating tags and converting to vector

```python
data['tags'] = data['product_category_tree']+ data['brand']+data['description']
data['tags']=data['tags'].apply(lambda x: " ".join(x))
data['tags'] = data['tags'].apply(lambda x: x.lower())
data['tags'][9008]
```

```python
data.head(0)
```

```python
## new DataFrame
all_products = data[['uniq_id','product_name','product_url','image', 'tags']]
all_products.head()
```

```python
products = all_products[:5000]
products.shape
```

# making vectors:

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps =  PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
products['tags'] = products['tags'].apply(stem)
# Stemming is used in information retrieval systems like search engines.
# It is used to determine domain vocabularies in domain analysis.
products.shape
```

```python
products.shape
```

```python
all_products['tags'][0]
```

```python
import re
def no_number_preprocessor(tokens):
    r = re.sub('(\d)+', '', tokens.lower())
    # This alternative just removes numbers:
    # r = re.sub('(\d)+', '', tokens.lower())
    return r
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english', preprocessor=no_number_preprocessor)
```

```python
vectors = cv.fit_transform(products['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
vectors.shape
```

```python
# len(data['tags'].unique())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity
```

```python
def recommend(product):
    product_index = products[products['product_name'] == product].index[0]
    distances = similarity[product_index]
    movie_list = sorted(list(enumerate(distances)), reverse= True, key = lambda x: x[1])[1:11]
    print(movie_list)
    for i in movie_list:
        print(products.iloc[i[0]].product_name)
```

```python
recommend("FabHomeDecor Fabric Double Sofa Bed")
```
